```python
#A bar plot of the 'branch' variable
plt.figure(figsize = (8, 4))
data.branch.value_counts().plot(kind = 'bar', rot = 0, color = ['red', 'green', 'blue'])
plt.xlabel('Branch')
plt.title('Distribution of Branches')
plt.show()
```

```python
#  A bar plot of the 'customer_type' variable
plt.figure(figsize = (4, 6))
data.customer_type.value_counts().plot(kind = 'bar', rot = 0, color = ['red', 'green'])
plt.xlabel('Customer Type')
plt.title('Distribution of Customer Type')
plt.show()
```

```python
# get the value counts of the 'customer_type' variable
data.customer_type.value_counts()
```

```python
# get the value counts of the 'gender' variable
data.gender.value_counts()
```

```python
# A bar plot of the 'product_line' variable
plt.figure(figsize = (8, 4))
data.product_line.value_counts().plot(kind = 'bar', rot = 15, 
                                   color = sns.color_palette('rocket', data.product_line.nunique()))
plt.xlabel('Product Line')
plt.title('Distribution of Product Line')
plt.xticks(ha = 'right')
plt.show()
```

```python
#A pie chart of the 'payment' variable
plt.figure(figsize = (8, 6))
data.payment.value_counts().plot(kind = 'pie', autopct = '%1.1f%%')
plt.title('Distribution of Method of Payment')
plt.show()
```

```python
# Heatmap of the correlations of the variables in the dataset
corr = data.corr()
fig, ax = plt.subplots(figsize = (10, 10))
colormap = sns.diverging_palette(220, 10, as_cmap = True)
sns.heatmap(corr, cmap = colormap, annot = True, fmt = ".2f")
plt.xticks(range(len(corr.columns)), corr.columns)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.show()
```

```python
# A pivot table to find the mean of gross income per branch
import numpy as np
data.head()
pd.pivot_table(data, index=['branch'], aggfunc={'gross_income': np.mean})
```

```python
# A pivot table to find the average gross income per product line
pd.pivot_table(data, index=['product_line'], aggfunc={'gross_income': np.average})
```

```python
#The average gross income per product line for each branch
branches = pd.pivot_table(data, index=['branch', 'product_line'], aggfunc={'gross_income' : 'mean'})
branches = branches.reset_index()
branches
```

```python
# The average gross income per product line for Branch A
branch_a = branches.loc[branches['branch'] == 'A']
branch_a
```

```python
# The average gross income per product line for Branch B
branch_b = branches.loc[branches['branch'] == 'B']
branch_b
```

```python
#The average gross income per product line for Branch C
branch_c = branches.loc[branches['branch'] == 'C']
branch_c
```
